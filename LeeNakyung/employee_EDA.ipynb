```python
import math

# 박스플롯(Boxplot)으로 시각화
# 수치형 변수만 추출
numeric_cols = employee_df.select_dtypes(include='number').columns

# 박스플롯 그리기 (4x4 격자)
plt.figure(figsize=(20, 15))

n = len(numeric_cols)
rows = math.ceil(n / 4)

plt.figure(figsize=(20, rows * 4))

for i, col in enumerate(numeric_cols):
    plt.subplot(rows, 4, i + 1)
    sns.boxplot(data=employee_df, y=col, color='skyblue')
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# IQR 방법으로 이상치 개수 확인
numeric_cols = employee_df.select_dtypes(include='number').columns

for col in numeric_cols:
    Q1 = employee_df[col].quantile(0.25)
    Q3 = employee_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = ((employee_df[col] < lower) | (employee_df[col] > upper)).sum()
    print(f'{col:<25} 이상치 수: {outlier_count}')
```

```python
# 이상치 많은 주요 변수 리스트 (숫자형 변수 기준)
outlier_vars = [
    'MonthlyIncome', 'NumCompaniesWorked', 'PerformanceRating', 'StockOptionLevel',
    'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany',
    'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager'
]

plt.figure(figsize=(14, 10))
sns.boxplot(data=employee_df[outlier_vars], orient='h')
plt.title('Boxplots of Variables with Many Outliers')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=employee_df)
plt.title('MonthlyIncome by Attrition')
plt.show()
```

```python
# 라벨 인코딩 (Attrition만 이진으로 먼저 변환)
employee_df['Attrition'] = employee_df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
employee_df.describe(include='all') # 숫자형 + 문자형
```

```python
# 수치형 변수 - 이직 여부에 따른 분포
num_cols = employee_df.select_dtypes(include=['int64', 'float64']).columns.drop('Attrition')

for col in num_cols:
    plt.figure()
    sns.kdeplot(data=employee_df, x=col, hue='Attrition', fill=True)
    plt.title(f"{col} vs Attrition")
    plt.show()
```

```python
# 범주형 변수 - 이직 여부에 따른 비율
cat_cols = employee_df.select_dtypes('object').columns

for col in cat_cols:
    plt.figure()
    sns.countplot(data=employee_df, x=col, hue='Attrition')
    plt.xticks(rotation=45)
    plt.title(f"{col} vs Attrition")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# 상관관계 히트맵
# -------------------------------
# Label Encoding (전체 object 컬럼)
for col in employee_df.select_dtypes('object').columns:
    employee_df[col] = LabelEncoder().fit_transform(employee_df[col])

corr = employee_df.corr()
plt.figure(figsize=(15, 12))
sns.heatmap(corr, annot=True, fmt=".2f", square=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Attrition 상관계수 Top 10
top_corr = corr['Attrition'].drop('Attrition').sort_values(key=abs, ascending=False)
top_corr.head(10)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=top_corr.head(10).values,
            y=top_corr.head(10).index)
plt.title("Top 10")
plt.xlabel("corr")
plt.ylabel('column')
plt.show()
```
